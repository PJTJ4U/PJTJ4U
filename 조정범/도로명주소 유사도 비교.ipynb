```python
from tqdm.notebook import tqdm
juso_api_key = 'C:\\Workspace\\python\\Data_Science\\04.공공데이터활용\\roadapikey.txt'
import pandas as pd
with open(juso_api_key) as f:
    api_key = f.read()
import requests
from urllib.parse import quote
road_url = 'https://www.juso.go.kr/addrlink/addrLinkApi.do'

addr_list = []
df_jibun['도로명주소'] = [0 for _ in range(len(df_jibun['지번주소']))]
n = 0
for i in tqdm(range(len(df_jibun['지번주소']))):
# for i in range(100):
    try:
        option = f'confmKey={api_key}&currentPage=1&countPerPage=10&keyword={quote(df_jibun["지번주소"][i])}' # f스트링에서 &을 통해 연결
        url = f'{road_url}?{option}&resultType=json'
        result = requests.get(url).json()
        addr = result['results']['juso'][0]['roadAddr']    
        df_jibun['도로명주소'][i] = addr
    except:
        n += 1
        print(i, n)
        df_jibun['도로명주소'][i] = '검색실패'
df_jibun
```
